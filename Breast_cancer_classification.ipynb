```python
from sklearn.model_selection import train_test_split
```

```python
X = df_cancer.drop(['target'],axis=1)

y = df_cancer['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train )
```

```python
y_predict = svc_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```

**Clearly it is a bad model , so let us improve our model by applying normalization and some hyperparameters**

```python
min_train = X_train.min()

max_train = X_train.max()

range_train = max_train - min_train

X_Scaled_train = (X_train - min_train)/range_train
```

```python
sns.scatterplot(x=X_Scaled_train['mean area'],y =X_Scaled_train['mean smoothness'],hue =y_train)
```

```python
sns.scatterplot(x=X_train['mean area'],y =X_train['mean smoothness'],hue =y_train)
```

```python
smodel = svc_model.fit(X_Scaled_train,y_train)
```

```python
X_
```

```python
min_test = X_test.min()

max_test = X_test.max()

range_test = max_test - min_test

X_Scaled_test = (X_test - min_test)/range_test
```

```python
y_predict = smodel.predict(X_Scaled_test)
```

```python
cmtest = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cmtest,annot=True)
```

**Now our model is working good **

```python
print(classification_report(y_test,y_pred=y_predict))
```

**IMPROVING MODEL BY CHECKING PARAMETERS LIKE C AND GAMMA**

```python
param_grid = {'C':[0.1,1.0,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=6)
```

```python
grid.fit(X_Scaled_train,y_train)
```

```python
grid.best_params_
```

```python
##trying with this best parameter

SVC_best_model = SVC(C=10,gamma=0.1,kernel='rbf')
```

```python
SVC_best_model.fit(X_Scaled_train,y_train)
```

```python
y_new_pred =SVC_best_model.predict(X_Scaled_test)
```

```python
confusion_matrix(y_test,y_new_pred)
```
